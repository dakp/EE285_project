```python
import csv
testaccu_1 = csv.reader(open('testaccu_AlexNet.csv'))
testaccu_2 = csv.reader(open('testaccu_ResNet.csv'))
testaccu_3 = csv.reader(open('testaccu_VGG16.csv'))
testaccu_AlexNet=[]
testaccu_ResNet=[]
testaccu_VGG16=[]
for i in testaccu_1:
    i = map(float, i)
    testaccu_AlexNet.extend(i)
# print(testaccu_AlexNet)
for i in testaccu_2:
    i = map(float, i)
    testaccu_ResNet.extend(i)
# print(testaccu_ResNet)
for i in testaccu_3:
    i = map(float, i)
    testaccu_VGG16.extend(i)
# print(testaccu_VGG16)
```

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
np.array(testaccu_AlexNet)
np.array(testaccu_ResNet)
np.array(testaccu_VGG16)
```

```python
plt.figure
x = range(len(testaccu_AlexNet))
plt.xlabel('Epoch')
plt.ylabel('Accuracy*100')
plt.ylim(40, 100)
plt.title('Test Accuracy for Different Models')
l1, = plt.plot(x,testaccu_AlexNet,"b",label = 'AlexNet')
l2, = plt.plot(x,testaccu_ResNet,"r",label = 'ResNet')
l3, = plt.plot(x,testaccu_VGG16,"g",label = 'VGG16')
plt.legend(handles = [l1, l2, l3,], labels = ['AlexNet','ResNet','VGG16'], loc = 'best')
plt.show()
plt.savefig("testaccu.png")
```
